```python
import os
import cv2
import json
import numpy as np
from tensorflow.keras.applications.resnet50 import preprocess_input
data_dir = '/home/elie.poncet@Digital-Grenoble.local/Documents/ML5/ML5/data/archive'

train_dir = os.path.join(data_dir, "train")
test_dir = os.path.join(data_dir, "valid")


train_annotation_file = '/home/elie.poncet@Digital-Grenoble.local/Documents/ML5/ML5/data/archive/train_annotations.txt'
test_annotation_file = '/home/elie.poncet@Digital-Grenoble.local/Documents/ML5/ML5/data/archive/valid_annotations.txt'

# Load the train annotations
with open(train_annotation_file, "r") as f:
    train_annotations = f.read().splitlines()

# Load the annotations
with open(test_annotation_file, "r") as f:
    annotations = json.load(f)

# Process the annotations
train_data = []
for annotation in annotations:
    image_id = annotation["image_id"]
    category_id = annotation["category_id"]
    bbox = annotation["bbox"]
    area = annotation["area"]
    segmentation = annotation["segmentation"]
    is_crowd = annotation["iscrowd"]


# Process the train data
train_labels = []
for annotation in annotations:
    image_id = annotation["image_id"]
    category_id = annotation["category_id"]
    bbox = annotation["bbox"]

    # Construct the image file name using the desired format
    image_file = f"image_id_{image_id:03}.jpg"
    image_path = os.path.join(train_dir, image_file)
    # Read the image using OpenCV
    image = cv2.imread(image_path)
  #
    resized_image = cv2.resize(image, (128, 128))
    normalized_image = preprocess_input(resized_image)
    train_data.append(normalized_image)
    train_labels.append(category_id)

# Convert the lists to NumPy arrays
train_data = np.array(train_data)
train_labels = np.array(train_labels)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
# Convert labels to integers
label_encoder = LabelEncoder()
train_labels_encoded = label_encoder.fit_transform(train_labels)

# Convert integers to one-hot vectors
onehot_encoder = OneHotEncoder(sparse=False)
train_labels_onehot = onehot_encoder.fit_transform(train_labels_encoded.reshape(-1, 1))
```

```python
import tensorflow as tf
from tensorflow.keras import layers
# Resize images to 64x64


model = tf.keras.Sequential([
    layers.Conv2D(16, (3, 3), activation='relu', input_shape=(128, 128, 3)),
    layers.MaxPooling2D(pool_size=2),
    layers.Conv2D(32, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=2),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=2),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(pool_size=2),
    layers.Dropout(0.4),
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.BinaryCrossentropy(),
              metrics=['accuracy'])




```

```python
model.summary()
```

```python

from tensorflow.keras.callbacks import EarlyStopping

# Define the early stopping callback
early_stopping = EarlyStopping(monitor='val_accuracy', patience=5)

# Train the model with early stopping
history = model.fit(train_data, train_labels, epochs=20000,batch_size=5,
                    validation_split=0.1)
```

```python
import matplotlib.pyplot as plt

training_acc = history.history['accuracy']
validation_acc = history.history['val_accuracy']

epochs = range(1, len(validation_acc) + 1)
plt.plot(epochs, validation_acc, 'b', label='Validation acc')
plt.plot(epochs, training_acc, 'r', label='Training acc')
plt.title( 'Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Acc')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

training_loss = history.history['loss']
validation_loss = history.history['val_loss']

epochs = range(1, len(validation_loss) + 1)
plt.plot(epochs, validation_loss, 'b', label='Validation Loss')
plt.plot(epochs, training_loss, 'r', label='Training loss')
plt.title( 'Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
